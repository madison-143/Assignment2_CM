# Linear Regression Analysis in R
This notebook demonstrates a simple linear regression analysis using R to model Salary based on Years of Experience.

Define the dataset

```r
dataset <- read.csv("regression_data.csv")
```

Create scatter plot (I changed the color)

```r
plot(dataset$YearsExperience, dataset$Salary, col = "green")
```

Fit the linear regression model and define it as "model"

```r
model <- lm(Salary ~ YearsExperience, data=dataset)
```

Load ggplot2 so we can work with it

```r
library(ggplot2)
```

Overlay regression line (I changed the colors again). 

geom_point creates a scatterplot by defining the dataset and assigning x and y, and I added the color at the end. 

geom_line adds the regression line to the scatterplot, and color is at the end again. 

ggtitle, xlab, and ylab are self explanatory.

```r
ggplot() +
geom_point(aes(x = dataset$YearsExperience, y = dataset$Salary), col = 'green') +
geom_line(aes(x = dataset$YearsExperience, y = predict(model, newdata = dataset)), col = 'purple') +
ggtitle('Salary vs Experience') +
xlab('Years of experience') +
ylab('Salary')
```

Evalute model!!

```r
summary(model)
```
