# Linear Regression
This notebook demonstrates a linear regression analysis using Python to model Salary based on Years of Experience.

Import pandas as pd, then define the dataset

```python
import pandas as pd
```

```python
dataset = pd.read_csv("regression_data.csv")
```

Import matplotlib.pyplot as plt and create a scatterplot

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(dataset["YearsExperience"], dataset["Salary"], color = "red")
```

This imports LinearRegression from the scikit-learn library

```python
from sklearn.linear_model import LinearRegression
```

Defines "model" and fits linear regression model to dataset

```python
model = LinearRegression()
model.fit(dataset[["YearsExperience"]], dataset[["Salary"]])
```

# Overlay regression line (changed colors)
plt.scatter - creates scatterplot by defining x and y columns, assigning color, and labeling it.
plt.plot - creates linear regression model based on model.predict of the dataset.
plt.title - this is what my title will be.
plt.xlabel - this is the x label.
plt.ylabel - this is the y label.
plt.legend - creates a legend based on my labels earlier in script.
plt.show - shows it

```python
plt.scatter(dataset["YearsExperience"], dataset["Salary"], color = "magenta", label = "Data points")
plt.plot(dataset["YearsExperience"], model.predict(dataset[["YearsExperience"]]), color = "teal", label = "Regression line")
plt.title("Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

Evaluate the model (R-squared)

```python
model.score(dataset[["YearsExperience"]], dataset[["Salary"]])
```

# Try to add R-squared to graph

define what r2 is (it's a surprise tool that will help us later).

```python
r2 = model.score(dataset[["YearsExperience"]], dataset[["Salary"]])
```

plt.scatter - create scatterplot

plt.plot - linear regression model

plt.annotate - tells python to create text of the r2 value as defined above. min and max tell python where to put the text. fontsize is self explanatory. color is self explanatory

plt. title, plt.xlabel, plt.ylabel, plt.show are the same as above

```python
plt.scatter(dataset["YearsExperience"], dataset["Salary"], color = "magenta", label = "Data points")
plt.plot(dataset["YearsExperience"], model.predict(dataset[["YearsExperience"]]), color = "teal", label = "Regression line")
plt.annotate(f'$R^2 = {r2:.3f}$', 
             xy = (dataset["YearsExperience"].min()*2.5, dataset["Salary"].max()*0.7), fontsize = 12, color = "purple")
plt.title("Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
