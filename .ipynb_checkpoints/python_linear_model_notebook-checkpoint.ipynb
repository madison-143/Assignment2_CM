# Linear Regression
This notebook demonstrates a linear regression analysis using Python to model Salary based on Years of Experience.

Import necessary packages and define the dataset

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from scipy.stats import linregress
```

```python
from sklearn.metrics import mean_squared_error
```

```python
dataset = pd.read_csv("regression_data.csv")
```

Define X and Y variables so I don't have to write it out every time

```python
X = dataset[["YearsExperience"]]
```

```python
Y = dataset[["Salary"]]
```

Fit the linear regression model to the data

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
model.score(X, Y)
```

Calculate slope and print value. slope.item converts scalar into normal float. The slope is 8285, and slope represents the estimated change in y for increases in x. A slope of 8285 means for each additional year of experience, the estimated salary increases by $8,285.

```python
slope = model.coef_[0]
slope = slope.item()
print(slope)
```

Calculate y-intercept and print value. intercept.item converts scalar into normal float. y-intercept is 29203, which means at 0 years of experience, the salary is estimated at $29,203.

```python
intercept = model.intercept_
intercept = intercept.item()
print(intercept)
```

Calculate prediced y values for any value of x - for line of best fit.

```python
y_predicted = model.predict(X)
```

MSE measures average squared error between predictions and actual values. Smaller MSE is better. In this example, the MSE is large, indicating that it's not great fit.

```python
mse = mean_squared_error(Y,y_predicted)
print(mse)
```

Calculate and print r2 and r values. Values closer to 1 for both r and r2 indicate better fit, and these values indicate a poorer fit.

```python
r2 = model.score(X, Y)
r = dataset['YearsExperience'].corr(dataset['Salary'])
print(r2)
print(r)
```

Graph the datapoints and fitted line, and add annotation, legend, and title/x and y-axis labels.

```python
plt.scatter(X, Y, color = "magenta", label = "Data points")
plt.plot(X, y_predicted, color = "teal", label = "Fitted line")
metrics_text = (
    f"y = {slope:.3f}x + {intercept:.3f}\n"
    f"r2 = {r2:.3f}\nr = {r:.3f}\nMSE = {mse:.3f}")
plt.text(1, 55000, metrics_text, fontsize = 10, color = "purple")
plt.title("Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
