# Linear Regression Analysis in R
This notebook demonstrates a simple linear regression analysis using R to model Salary based on Years of Experience.

define dataset

```r
dataset <- read.csv("regression_data.csv")
```

define the linear regression model, slope, intercept, r2, and r values

```r
model <- lm(Salary ~ YearsExperience, data=dataset)
model_summary <- summary(model)
slope <- coef(model)[2]
intercept <- coef(model)[1]
r2 <- model_summary$r.squared
r <- cor(dataset$YearsExperience, dataset$Salary)
```

print values

```r
cat(
  "Slope:", slope, "\n",
  "Intercept:", intercept, "\n",
  "r2:", r2, "\n",
  "r:", r, "\n"
)
```

Define the mean squared error and print value

```r
mse <- mean((predict(model) - dataset$Salary)^2)
cat("MSE:", mse, "\n")
```

Load ggplot2 in for graphing abilities

```r
library(ggplot2)
```

Motherload of graph code. Additions include annotations: adding y=mx+b, r2, r, and MSE text to graph. Also, I used scale_color_manual to add a legend for funsies.

```r
ggplot(dataset, aes(x = YearsExperience, y = Salary)) +
geom_point(aes(color = "Data points")) +
geom_line(aes(y = predict(model, newdata = dataset), color = "Fitted line")) +
annotate("text", x = 2, y = 60000, col = "navy",
         label = paste(
            "y =", round(slope, 2),"x +", round(intercept,2), "\n",
            "r2:", round(r2, 2), "\n",
            "r:", round(r, 2), "\n",
            "MSE:", round(mse, 2), "\n")) +             
ggtitle('Salary vs Experience') +
xlab('Years of experience') +
ylab('Salary') +
scale_color_manual(
    name = "Legend",
    values = c("Data points" = "green", 
               "Fitted line" = "purple"))
```
